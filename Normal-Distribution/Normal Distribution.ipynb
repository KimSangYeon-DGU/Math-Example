```python
%matplotlib inline

import math
from matplotlib import pyplot as plt

def gauss_distf(x, mu=0, sigma=1):
    return (math.exp((-(x-mu)**2)/(2*(sigma**2))))/(math.sqrt(2*math.pi*(sigma**2)))
```

```python
xs = [x/10.0 for x in range(-50, 50)]

plt.plot(xs, [gauss_distf(x, sigma=0.5) for x in xs], ':', label="mu=0, sigma=0.5")
plt.plot(xs, [gauss_distf(x, sigma=1) for x in xs], '-', label="mu=0, sigma=1")
plt.plot(xs, [gauss_distf(x, sigma=2) for x in xs], '--', label="mu=0, sigma=2")
plt.legend()
plt.title("Normal distibution")
```

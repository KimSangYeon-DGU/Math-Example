```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import math

def logistic_function(x):
    return 1/(1+math.exp(-x))

def constant_function(x):
    return x

_domain = np.arange(-5, 5, 0.1)
_range = [logistic_function(x) for x in _domain]
_constant = [constant_function(0.5) for _ in _domain]

plt.figure(figsize=(12,8))
plt.title("Logistic Function", fontsize=20)

plt.plot(_domain, _range, color="red")
plt.plot(_domain, _constant, linestyle=':', color="grey")

plt.legend([r"$y=\frac{1}{1+e^{-x}}$", "$y = 0.5$"], prop={'size':25})
plt.show()
```
